```python
df = pd.read_csv('table(1).csv', delimiter=';', decimal=',')

# Convert the 'Tid(norsk normaltid)' column to datetime
df['Tid(norsk normaltid)'] = pd.to_datetime(df['Tid(norsk normaltid)'], format='%d.%m.%Y')

# Set the date column as the index
df.set_index('Tid(norsk normaltid)', inplace=True)

# Verify the DataFrame (Optional)
print(df.head())
```

```python
# Resample to get the first date of each month
monthly_data = df.resample('MS').first()

# Reset the index to use the Date column for plotting
monthly_data.reset_index(inplace=True)

monthly_data
```

```python
# Matplotlib plot
plt.figure(figsize=(10, 6))
plt.plot(monthly_data['Tid(norsk normaltid)'], monthly_data['Middeltemperatur (døgn)'], label='Middeltemperatur', marker='o', linestyle='-')
plt.title('Middeltemperatur on the First Date of Each Month')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()

# Seaborn plot for enhanced visual
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Tid(norsk normaltid)', y='Middeltemperatur (døgn)', data=monthly_data, label='Middeltemperatur', marker='o')
plt.title('Middeltemperatur on the First Date of Each Month')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
temp_data = monthly_data['Middeltemperatur (døgn)'][:-1].to_numpy()

yearly_averages = []
sd_year = []

for i in range(0, len(temp_data), 12):
    yearly_averages.append(np.sum(temp_data[i: i+12])/12)
    sd_year.append(np.std(temp_data[i: i+12], ddof=1))

print(yearly_averages)
print(sd_year)
```

```python
plt.plot(np.arange(len(yearly_averages)), yearly_averages)
```

```python
# Step 3: Resample to yearly frequency and calculate the mean
yearly_data = df.resample('YE').mean()
yearly_data.reset_index(inplace=True)
yearly_data.rename(columns={'Tid(norsk normaltid)': 'Year'}, inplace=True)

# Step 4: Create yearly average plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(yearly_data['Year'], yearly_data['Middeltemperatur (døgn)'], label='Average Temperature', marker='o', linestyle='-')
plt.title('Yearly Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Create yearly average plot using Seaborn
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Middeltemperatur (døgn)', data=yearly_data, label='Average Temperature', marker='o')
plt.title('Yearly Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```
